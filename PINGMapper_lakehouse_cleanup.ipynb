# Cleanup — delete the ingested sonar batch from VForce Lakehouse

A companion to `PINGMapper_lakehouse_flow.ipynb`. Lists batches, deletes the one we ingested (objects + gold rows + queue jobs), and shows the admin purge for a full reset.


```python
import os, requests
LAKEHOUSE_URL=os.environ.get('LAKEHOUSE_URL','https://lakehouse.vforce360.ai')
TOKEN=os.environ.get('LAKEHOUSE_TOKEN',''); TENANT=os.environ.get('LAKEHOUSE_TENANT','default')
H={'X-Tenant-Id':TENANT}
if TOKEN: H['Authorization']='Bearer '+TOKEN
```

### Find the batch


```python
r=requests.get(LAKEHOUSE_URL+'/api/v1/batches', headers=H)
print(r.status_code); print(r.text[:800])
# Copy the batch id you want to remove (the 'sonar-...' batch from the ingest notebook):
BATCH_ID=os.environ.get('BATCH_ID','')  # e.g. 'sonar-Test-Small-DS-YYYYMMDD-HHMM'
```

### Delete that batch (scoped: objects + gold rows + jobs)


```python
if BATCH_ID:
    r=requests.delete(LAKEHOUSE_URL+'/api/v1/batches/'+BATCH_ID, headers=H)
    print(r.status_code, r.text[:300])
else:
    print('Set BATCH_ID first.')
```
